```python
# numpy27.py
import numpy as np
# Compute outer product of vectors
v = np.array([1,2,3])
w = np.array([4,5])
# reshape v to be a column vector of shape (3, 1); then broadcast it against w to an output of shape (3, 2),
print(np.reshape(v, (3, 1)) * w)
# Add a vector to each row of a matrix
x = np.array([[1,2,3], [4,5,6]])
# x has shape (2, 3) and v has shape (3,) so they broadcast to (2, 3),
print(x + v)
```

```python
# matplotlib31b.py
import numpy as np
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana','Cherry', 'Durian', 'Grape']
np.random.seed(1)
sales2020 = np.random.randint(10, 50, 5)
np.random.seed(2021)
sales2021 = np.random.randint(10, 50, 5)
plt.bar(fruits, sales2020, color='r')
plt.bar(fruits, sales2021, bottom=sales2020, color='b')
plt.xticks(fruits)
plt.legend(labels=['2020','2021'])
plt.show()
```

```python
# matplotlib35.py
from matplotlib import pyplot as plt
import numpy as np
fruits = ['Apple', 'Banana', 'Cherry', 'Durian', 'Grape']
np.random.seed(1)
sales = np.random.randint(10, 50, 5)
plt.pie(sales, labels=fruits, autopct='%1.2f%%')
plt.title('2020 sales report')
plt.show()
```
