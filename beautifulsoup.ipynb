```python
## robots01.txt
from urllib import robotparser
robot_parser = robotparser.RobotFileParser()
def prepare(robots_txt_url):
    robot_parser.set_url(robots_txt_url)
    robot_parser.read()
def is_allowed(target_url, user_agent='*'):
    return robot_parser.can_fetch(user_agent, target_url)
prepare('https://twitter.com/robots.txt')
print(is_allowed('https://twitter.com/covers/'))
print(is_allowed('https://twitter.com/python/'))
print(is_allowed('https://twitter.com/search/realtime/'))
```

```python
## scraping11.py
from urllib.request import urlopen, urljoin
import re
def download_page(url):
    return urlopen(url).read().decode('utf-8')
def extract_links(page):
    link_regex = re.compile('<a[^>]+href=["\'](.*?)["\']',re.IGNORECASE)
    return link_regex.findall(page)
target_url = 'https://twitter.com/'
site = download_page(target_url)
links = extract_links(site)
for link in links:
    print(link)
for link in links:
    print(urljoin(target_url, link))
# scraping15.py
def extract_image_locations(page):
    img_regex = re.compile('<img[^>]+src=["\'](.*?)["\']',re.IGNORECASE)
    return img_regex.findall(page)
image_locations = extract_image_locations(site)
for src in image_locations:
    print(urljoin(target_url, src))
```

```python
# scraping20.py
import requests
def save_page(url, save_file):
    htmlfile = requests.get(url)
    with open(save_file, 'wb') as file_obj:
        for diskStorage in htmlfile.iter_content(40960):
            size = file_obj.write(diskStorage)
            print(size)
        return True
    return False

target_url = 'https://www.facebook.com/'
save_file = 'save_page.html'
result = save_page(target_url, save_file)
if result:
    print(save_file)
```
