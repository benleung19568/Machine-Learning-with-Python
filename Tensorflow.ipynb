```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
layer0 = Dense(units=1, input_shape=[1])
model = Sequential([layer0])
model.compile(optimizer='sgd', loss='mean_squared_error')
X = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X,y) 
```

```python
model.fit(X, y, epochs=500)
# Predict result
print(model.predict([0.5]))
print(model.predict([10.0]))
```

```python
model = Sequential([
    Conv2D(filters=16,
           kernel_size=(3,3),
           activation='relu',
           input_shape=(28,28,1)),
    MaxPooling2D((3,3)),
    Flatten(),
    Dense(10, activation='softmax')
])
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.05)
loss = tf.keras.losses.SparseCategoricalCrossentropy()
acc = tf.keras.metrics.SparseCategoricalAccuracy()
mae = tf.keras.metrics.MeanAbsoluteError()

model.compile(
    optimizer=opt,
    loss=loss,
    metrics=[acc,mae]
)

print(model.optimizer)
print(model.loss)
print(model.metrics)
print(model.optimizer.lr)
```

```python
history = model.fit(train_images, train_labels, epochs=8, batch_size=256)
```

```python
row_number =3
column_number = 3

np.random.seed(10)
random_list = np.random.randint(0,
                                img_train.shape[0],
                                int(row_number * column_number)
                                )
print(random_list)
```

```python
plt.figure(figsize=(6,6))
for i, random_number in enumerate(random_list):
  label = label_train[random_number]
  plt.subplot(column_number, row_number, i+1)
  plt.axis('off')
  plt.title("{}".format(label))
  plt.imshow(img_train[random_number], cmap='gray')
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='Softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(img_train, label_train, epochs=5)
```

```python
model.evaluate(img_test, label_test)
```
